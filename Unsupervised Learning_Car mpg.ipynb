# DATA CLUSTERING

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
scaled_df=scale.fit_transform(imputed_df)
```

```python
from sklearn.cluster import KMeans,AgglomerativeClustering
kmeans=KMeans()
kmeans.fit(scaled_df)
```

```python
kmeans.labels_
```

```python
import numpy as np
wss =[] 
K=np.arange(1,21)
K
for k in K:
    k_means=KMeans(n_clusters=k)
    k_means=k_means.fit(scaled_df)
    wss.append(k_means.inertia_)
    print("No. of Clusters:",k ,"WSS:",k_means.inertia_)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(K,wss,'bx-')
plt.xlabel("No.Of Clusters")
plt.ylabel('Sum of Squares')
```

Optimum number of clusters using elbow method turns out to be 4. Beyond 4 there is not much reduction in errors.

```python
k_means=KMeans(n_clusters=4)
k_means=k_means.fit(scaled_df)
print(k_means.inertia_)
```

```python
from sklearn.metrics import silhouette_samples, silhouette_score
silhouette_score(scaled_df, labels=k_means.labels_)
```

```python
hierarchical=AgglomerativeClustering(linkage='ward')
hierarchical=hierarchical.fit(scaled_df)
```

```python
hierarchical.labels_
```

```python
hierarchical.children_
```

```python
from scipy.cluster.hierarchy import dendrogram,linkage
Z=linkage(scaled_df,method='ward')
```

```python
plt.figure(figsize=(10, 10))
plt.title('Dendogram')
plt.xlabel('Index')
plt.ylabel('Distance')
dendrogram(Z, leaf_rotation=90)
plt.tight_layout()
```

Difference Between KMeans and Hierarchical Approach:
    
1. KMeans requires prior knowledge on number of clusters(either using elbow/silhouette scores) whereas in Hierarchical we can interpret the number of clusters using the dendrogram diagram on the go.
2. KMeans randomly picks centroid and due to this the clusters are bound to change whereas in Hierarchical approach, for a given linkage method(ward/single/complete/average) the clusters remain the same.

# Optimal Clusters

Optimal number of clusters in this case comes out to be 4 as depicted using elbow diagram. Also Dendrogram diagram shows the number of longest branch(longer length indicates number of clusters different from each other, in this case 4 lines are relatively longer when cut through at distance of around 15)  turns out to be 4. 

```python
from scipy.cluster.hierarchy import fcluster
clusters = fcluster(Z, t=15, criterion='distance')
clusters
```

```python
new_df=pd.DataFrame(scaled_df)
new_df.columns=imputed_df.columns
new_df["clusters"]=clusters
new_df
```

```python
new_df_1=new_df[new_df["clusters"]==1].drop(columns="clusters")
new_df_2=new_df[new_df["clusters"]==2].drop(columns="clusters")
new_df_3=new_df[new_df["clusters"]==3].drop(columns="clusters")
new_df_4=new_df[new_df["clusters"]==4].drop(columns="clusters")
df_array=[new_df_1,new_df_2,new_df_3,new_df_4]
new_df_1
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
X_train, X_test, y_train, y_test=train_test_split(new_df.drop(columns=["mpg","clusters"]),new_df["mpg"],random_state=1,test_size=0.30)
linear_model=LinearRegression()
linear_model.fit(X_train,y_train)
print("Intercept: ",linear_model.intercept_,"Co-Efficient: ",linear_model.coef_)
y_pred = linear_model.predict(X_test)
result = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
print("Mean Absolute Error : ",metrics.mean_absolute_error(y_test,y_pred),"Mean Squared Error : ",metrics.mean_squared_error(y_test,y_pred))
print()
result
```

```python
results=[]
for cluster in df_array:
    X_train, X_test, y_train, y_test=train_test_split(cluster.drop(columns="mpg"),cluster["mpg"],random_state=1,test_size=0.30)
    linear_model=LinearRegression()
    linear_model.fit(X_train,y_train)
    print("Intercept: ",linear_model.intercept_,"Co-Efficient: ",linear_model.coef_)
    y_pred = linear_model.predict(X_test)
    results.append(pd.DataFrame({'Actual': y_test, 'Predicted': y_pred}))
    print("Mean Absolute Error : ",metrics.mean_absolute_error(y_test,y_pred),"Mean Squared Error : ",metrics.mean_squared_error(y_test,y_pred))
    print()
```

Using different models for different clusters helps to improve the accuracy and prediction, as the data points in a particular cluster are similar to each other and follows a certain pattern.

Also unsupervised learning helps unearths hidden pattern in the given dataset so that data points within clusters are similar within clusters and are different between data points in a cluster. Naturally, building models on datapoints which has already been clustered or grouped helps improve accuracy. Where as, using single model to predict the entire datasets are prone to error.

In short, using multiple models helps build a better model.

In addition to the above, performance wise unsupervised learnings are computationally expensive, as it is an iterative process.
