The regularization parameters should be tuned to find a balance between variance and bias errors.

```python
dt_regularized=DecisionTreeClassifier(criterion='entropy',min_samples_leaf=10,max_depth=3,random_state=10)
dt_reg_model=dt_regularized.fit(X_train,y_train)
train_reg_pred=dt_reg_model.predict(X_train)
test_reg_pred=dt_reg_model.predict(X_test)
print('Training accuracy: ',accuracy_score(y_train,train_reg_pred))
print('Testing accuracy: ',accuracy_score(y_test,test_reg_pred))
```

```python
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (3,3), dpi=300,facecolor='k')
tree.plot_tree(dt_regularized,feature_names = list(X_train), 
               class_names=list(class_labels),
               filled = True)
plt.show()
```

Regularization of these parameters(like min_samples_leaf,max_depth) helps us to find the balance between overfit and underfit model. 

That is, it helps to find a tradeoff between overfit and underfit model, where underfit model is too simple that the model has learnt very less on the training data and is prone to bias errors, whereas overfit model has learnt training data with maximum accuracy such that the model cannot work with test datas that are not seen before by the model leading to variance error.

In short, regularization parameters helps to find the balance between bias(underfit model) and variance(overfit model) errors either alone or in combination with other regualarization parameters.
