# QUESTION 4

```python
la_liga["Goal Difference"]=pd.to_numeric(la_liga["GoalsFor"])-pd.to_numeric(la_liga["GoalsAgainst"])
gd={}

def Goal_diff_count():
    for teams in range(la_liga.index.size):
        team=la_liga.index[teams]
        Goals_difference=la_liga.loc[team]["Goal Difference"]
        gd[team]=Goals_difference
    gd['max']=la_liga["Goal Difference"].idxmax()
    gd['min']=la_liga["Goal Difference"].idxmin()
    print("Team with max Goal Difference:",gd['max'],",Goal Difference",la_liga["Goal Difference"].max())
    print("Team with min Goal Difference:",gd['min'],",Goal Difference",la_liga["Goal Difference"].min())
    return gd

Goal_diff_count()

```

# QUESTION 5

```python
la_liga["GamesWon"]=pd.to_numeric(la_liga["GamesWon"])
la_liga["GamesPlayed"]=pd.to_numeric(la_liga["GamesPlayed"])
la_liga["Winning Percent"]=(la_liga["GamesWon"]/la_liga["GamesPlayed"])*100
la_liga["Winning Percent"].fillna(value=0,inplace=True)
la_liga
```

```python
la_liga_winning_percentage_df=la_liga.sort_values(by=['Winning Percent'],ascending=False)
la_liga_winning_percentage_df.head(5).index.to_list()
```
